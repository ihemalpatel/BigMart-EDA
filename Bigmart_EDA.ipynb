```python
df['Item_Weight'] = df['Item_Weight'].fillna(df['Item_Weight'].mean())
```

```python
df['Outlet_Size'] = df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0])
```

```python
df.head()
```

```python
df['Item_Fat_Content'].value_counts()
```

```python
df['Item_Fat_Content'].replace({'LF':'Low Fat','low fat':'Low Fat', 'reg':'Regular'} ,inplace = True)
```

```python
df['Item_Weight'] = round(df['Item_Weight'],1)
```

```python
df['Item_Identifier'] = df['Item_Identifier'].str[:2]
```

```python
df['Item_Visibility'] = round(df['Item_Visibility'],2)
```

```python
df['Item_MRP'] = round(df['Item_MRP'])
```

```python
df['Item_Outlet_Sales'] = round(df['Item_Outlet_Sales'])
```
